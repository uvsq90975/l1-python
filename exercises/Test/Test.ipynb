```python
# Test GUI IN100 janv 2021

import tkinter as tk 

WIDTH, HEIGHT = 600,600

racine = tk.Tk()

# Canvas

canvas = tk.Canvas(racine, width = WIDTH, height = HEIGHT, bg ="black")
canvas.grid(row = 0, column = 0 )


# Rectangles de couleurs

carre_b = canvas.create_rectangle(0, 0, 50, 50, fill ="blue")
carre_r = canvas.create_rectangle(50, 0, 100, 50, fill = "red")
carre_w = canvas.create_rectangle(100, 0, 150, 50, fill = "white")


# Cercle 

coul = ["yellow"]
cercle = canvas.create_oval((WIDTH/2 - 50, HEIGHT/2 - 50), (WIDTH/2 + 50, HEIGHT/2 + 50), fill = coul[-1])


def aff_cercle(event):
    global coul
    x, y = event.x, event.y
    print(x, y)
    if 0<x and x<50 and 0<y and y<50 :
        coul.append("blue")
    elif 50<x and x<100 and 0<y and y<50 :
        coul.append("red")
    elif 100<x and x<150 and 0<y and y<50 :
        coul.append("white")
    canvas.itemconfigure(cercle, fill=coul[-1])
    print(coul)


# Clic dans canvas

racine.bind("<Button-1>", aff_cercle)


# Boutton Annuler

def annuler():
    global coul
    if len(coul) > 1:
        del coul[-1]
        del coul[-1]
        canvas.itemconfigure(cercle, fill = coul[-1])
        print(coul)


bt_annuler = tk.Button(racine, text= "Annuler", bg="green", command = annuler)
bt_annuler.grid(row = 0, column = 1)


# Fin du code

racine.mainloop()
```

```python
import tkinter as tk

#Constantes
cpt = 0
L = 600
HEIGHT, WIDTH = L, L
x0, y0 = L//2 - 30, L//2 - 30
x1, y1 = L//2 + 30, L//2 + 30


#Fonctions
def dim_carre(event):
    global cpt, x0, y0, x1, y1
    if x0 < event.x and event.x < x1 and y0 < event.y and event.y < y1:
        x0 -= 10
        y0 -= 10
        x1 += 10
        y1 += 10
    else:
        x0 += 10
        y0 += 10
        x1 -= 10
        y1 -= 10
    canvas.delete(carre)
    carre = canvas.create_rectangle(x0, y0, x1, y1, fill="yellow")


def pause():
    global cpt 
    cpt += 1
    if cpt%2 != 0 :
        canvas.config(text="Restart")
    else :
        canvas.config(text="Pause")



#FenÃªtres tkinter (bouton,canvas...)
racine = tk.Tk()

canvas = tk.Canvas(racine, height=HEIGHT, width=WIDTH, bg="blue")
canvas.grid(row=0, column=0)

b_pause = tk.Button(racine, text="Pause",command=pause, font=("Helvetica","20"))
b_pause.grid(row=1, column=0)

canvas.create_rectangle(x0, y0, x1, y1, fill="yellow")

canvas.bind("<Button-1>", dim_carre)

racine.mainloop()
```
